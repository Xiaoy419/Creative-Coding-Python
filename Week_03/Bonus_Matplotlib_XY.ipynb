```python
import matplotlib.pyplot as plt
```

```python
price = {
    'day 1': 38, 
    'day 2': 33, 
    'day 3': 28, 
    'day 4': 25, 
    'day 5': 21, 
    'day 6': 18,
    'day 7': 23, 
    }
```

```python
plt.scatter(range(len(price)), list(price.values()), color=['#236E96', '#15B2D3', '#e2f4c7', '#eae374', '#FFD700', '#F3872F', '#FF598F'])
plt.xticks(range(len(price)), list(price.keys()), rotation = 45)
plt.show()
```
